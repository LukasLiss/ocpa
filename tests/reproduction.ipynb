```python
# Precision & Fitness
import ocpa
import ocel
print(ocpa.__version__)

# Converting XMLOCEL -> JSONOCEL
ocel_log = ocel.import_log("log.xmlocel")
ocel.export_log(ocel_log,'log.jsonocel')

df, _ = ocpa.objects.log.importer.ocel.factory.apply(f"log.jsonocel", parameters={'return_df': True})
ocel_obj = ocpa.objects.log.obj.OCEL(df)
print(ocel_obj)

ocpn = ocpa.algo.discovery.ocpn.algorithm.apply(df)
precision, fitness = ocpa.algo.evaluation.precision_and_fitness.evaluator.apply(ocel_obj,ocpn)
print(f"Precision: {precision}, Fitness: {fitness}")

```
